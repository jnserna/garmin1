```python
def tabla_guarda(df,date_cols=False):
  # Convierte columnas en str
  df.replace([np.inf, -np.inf], np.nan, inplace=True)

  num_cols = df.select_dtypes(include=['float64','Int64']).columns
  for i in num_cols:
    df[i].fillna(0, inplace=True)


  if date_cols==True:
    date_cols_DATE = df.select_dtypes(include=['datetime64']).columns
    try:
      for col in date_cols_DATE:
        df[col] = df[col].dt.strftime('%Y-%m-%d')
    except:
      pass

  df.fillna('', inplace=True)
```

```python
# Acá voy a romper el dict del activity type

# Extracting dictionary values from the 'Details' column and creating new columns
df_details = df2_1['activityType'].apply(lambda x: {} if pd.isna(x) else x).apply(pd.Series)[['typeKey']]

# Concatenating the new columns to the original DataFrame
df2_2 = pd.concat([df2_1, df_details], axis=1)

# Drop columns
df2_3 = df2_2.drop(columns=['activityType'])

#Rename col
df2_3 = df2_3.rename(columns={"typeKey": "activityType"})

# Sorting columns
df2_4 = df2_3[['activityName', 'activityType','startTimeLocal','locationName', 'duration','distance','calories', 'steps','vO2MaxValue','avgStrideLength','elevationGain','max20MinPower','description']]
```

```python
df2_4
```

```python
# Cambiar el formato de fecha a datetime
df2_4['startTimeLocal'] = pd.to_datetime(df2_4['startTimeLocal'], format='%Y-%m-%d %H:%M:%S')
df2_4['activity_date'] = df2_4['startTimeLocal'].dt.strftime('%Y-%m-%d')
df2_4
```

```python
# Convertir la duración (de segundos a horas)
df2_4['duration'] = pd.to_timedelta(df2_4['duration'], unit='S')
df2_4
```

```python
df2_4.info()
```

```python
df2_4['duration_hours'] = df2_4['duration'] / np.timedelta64(1, 'h')

df2_4.info()
```

```python

df2_4['distance_km'] = df2_4['distance'] / 1000
df2_4.info()
```

```python
# Drop columns
df2_5 = df2_4.drop(columns=['startTimeLocal', 'duration','distance'])
# Sorting columns
df2_5 = df2_5[['activityName', 'activityType','activity_date','locationName', 'duration_hours','distance_km','calories', 'steps','vO2MaxValue','avgStrideLength','elevationGain','max20MinPower','description']]
```

```python
df2_5.info()
```

```python
# Agregar una columna con mi propia clasificación (activity type 2)

def get_act_type2(activityType):
    if activityType == 'running':
        return '3-running'
    elif activityType == 'walking':
        return '2-walking'
    elif activityType == 'road_biking':
        return '1-cycling'
    elif activityType == 'cycling':
        return '1-cycling'
    elif activityType == 'virtual_ride':
        return '1-cycling'
    elif activityType == 'lap_swimming':
        return '4-lap_swimming'
    elif activityType == 'treadmill_running':
        return '3-running'
    elif activityType == 'indoor_cycling':
        return '1-cycling'
    elif activityType == 'indoor_cardio':
        return '5-others'
    elif activityType == 'other':
        return '5-others'
```
